```python
all_bluebike = pd.concat([bluebikes0119, bluebikes0219, bluebikes0319, bluebikes0419, bluebikes0519, bluebikes0619, bluebikes0719, bluebikes0819, bluebikes0919, bluebikes1019, bluebikes1119, bluebikes1219, bluebikes0120, bluebikes0220, bluebikes0320, bluebikes0420, bluebikes0520, bluebikes0620, bluebikes0720, bluebikes0820, bluebikes0920, bluebikes1020, bluebikes1120, bluebikes1220, bluebikes0121, bluebikes0221, bluebikes0321, bluebikes0421, bluebikes0521])
```

```python
all_bluebike.info()
```

```python
all_bluebike = all_bluebike[['bikeid', 'starttime', 'stoptime', 'start station id', 'end station id']]
all_bluebike.rename(columns={'bikeid':'ride_id', 'starttime':'started_at', 'stoptime':'ended_at', 'start station id':'start_station_id', 'end station id': 'end_station_id'}, inplace=True)
```

```python
all_bluebike['started_at'] = pd.to_datetime(all_bluebike['started_at'], format='%Y-%m-%d')
all_bluebike['ended_at'] = pd.to_datetime(all_bluebike['ended_at'], format='%Y-%m-%d')
```

```python
all_bluebike['duration'] = all_bluebike['ended_at'] - all_bluebike['started_at']
```

```python
all_bluebike = all_bluebike.dropna(subset=['start_station_id'])
all_bluebike = all_bluebike.dropna(subset=['end_station_id'])
all_bluebike = all_bluebike[~((all_bluebike['start_station_id'] == all_bluebike['end_station_id']) & (all_bluebike['duration'] < '00:05:00'))]
```

```python
all_bluebike['start_date'] = all_bluebike['started_at'].dt.date
```

```python
bluebike_by_date = all_bluebike.groupby('start_date').agg({'started_at':'count'}).reset_index()
bluebike_by_date.rename(columns={'start_date':'ride_date', 'started_at':'ride_count'}, inplace = True)
```

```python
bluebike_by_date['ride_date'] = pd.to_datetime(bluebike_by_date['ride_date'], format='%Y-%m-%d')
```

```python
bluebike_by_date.info()
```

```python
bluebike_by_date['month'] = bluebike_by_date['ride_date'].dt.month
bluebike_by_date['year'] = bluebike_by_date['ride_date'].dt.year
bluebike_by_date['m_y'] = bluebike_by_date['month'].apply(str) + '-' + bluebike_by_date['year'].apply(str) 
```

```python
bluebike_by_date.groupby(['year', 'month']).agg({'ride_count':'sum'})
```

```python
plt.barh('m_y', 'ride_count', data=bluebike_by_date)
plt.tight_layout()
```
